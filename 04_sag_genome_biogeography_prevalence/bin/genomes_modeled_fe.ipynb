# Download modeled surface dissolved Fe concentrations

Modeled Fe data comes from [Simons CMAP](https://simonscmap.com/). There is a very nice [python API here](https://cmap.readthedocs.io/en/latest/user_guide/API_ref/pycmap_api/pycmap_api_ref.html)

```python
import pycmap
import pandas as pd
import numpy as np
import xarray as xr
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
from cmocean import cm as cmo
from matplotlib import ticker, cm
```

```python
pycmap.__version__
```

To retrieve data, we need to create an instance of the system's API and pass the API key. It is no necessary to pass the API key every time you run a code locally, because it will be stored locally. The API class has other optional parameters to adjust its behavior. All parameters can be updated persistently at any point in the code.

```python
api = pycmap.API('9e812810-4e31-11ea-93ba-e9f0ed605cb0')
```

We need modeled surface Fe concentrations. We want them from the PISCES model

```python
api.search_catalog('Fe model')
```

OK found the variable. Now need to find the spatial/temporal range.

This was confusing because the model grid doesn't go all the way from -90 to +90 latitude. However, there is no way of getting the exact latitude range of the model from the API. Also if you ask for data from -90 to +90 latitude you get a SQL error rather than it just giving you the entire extent of avaialable latitudes. That's why here I am using the relatively arbitrary cutoff of -76 to +76 latitude. This is good enough for our purposes.

```python
table = 'tblDarwin_Nutrient_Climatology'
variable = 'FeT_darwin_clim'
dt1 = '2008-01-01'
dt2 = '2008-12-31'
lat1, lat2, lon1, lon2 = -76, 76, -180, 180
depth1, depth2 = 0, 55.1
```

Ok now the dataset is stored in a pandas dataframe format.

```python
df = api.space_time(table=table, variable=variable, dt1=dt1, dt2=dt2, 
               lat1=lat1, lat2=lat2, lon1=lon1, 
               lon2=lon2, depth1=depth1, depth2=depth2)
```

Save pandas dataframe

```python
compression_opts = dict(method='bz2',archive_name='../../shared_data/FeT_Darwin_0.50.monthly_v0.1_llc90.csv')  
df.to_csv('../../shared_data/FeT_Darwin_0.50.monthly_v0.1_llc90.bz2', index=False, compression=compression_opts)  
```

## Plot modeled Fe with genome biogeography

Start from here to load locally saved PISCES data

Note for future self: [Very handy chart](https://pandas.pydata.org/pandas-docs/stable/getting_started/comparison/comparison_with_r.html) for converting dplyr syntax to pandas

```python
df = pd.read_csv('../../shared_data/FeT_Darwin_0.50.monthly_v0.1_llc90.bz2')  
```

Concentrations are mmol/m^3 

```python
df.head()
```

Calculate depth and time average mean value

```python
df1 = df.groupby(['lat', 'lon'], as_index=False).mean().reset_index()[['lat', 'lon', 'FeT_darwin_clim']]
```

```python
api.get_unit('tblDarwin_Nutrient_Climatology', 'FeT_darwin_clim')
```

Log10 transform Fe concentrations. 

```python
df1['Fenorm'] = np.log10(df1['FeT_darwin_clim'])
```

Now convert to an xarray dataset

```python
ds = df1.set_index(['lat', 'lon']).to_xarray()
```

```python
xr.set_options(display_style="html")
ds
```

Look at dataset mins/maxs to get sense for colorbar limits

```python
ds.min()
```

```python
ds.max()
```

Load the siderophore pres/abs data that we want to plot

```python
bins = pd.read_csv("../../shared_data/pro_syn_metadata.tsv", delimiter='\t')
bins.head()
```

### Formatting prochlorococcus counts

```python
probins = bins[bins['GROUP'].isin(['prochlorococcus'])]
```

```python
probincounts = probins.groupby(['LATITUDE','LONGITUDE']).size().reset_index(name="bins").rename(
    columns={'LATITUDE':'lat', 'LONGITUDE':'lon'})
```

```python
# to make the circle size more visible in map
probincounts['bins'] = probincounts['bins'].clip(lower=4)
```

```python
prosidcounts = probins[['LATITUDE', 'LONGITUDE', 'SIDEROPHORE_TRANSPORTER_COUNT']].groupby(
    ['LATITUDE','LONGITUDE']).agg(
    siderophore_bins=('SIDEROPHORE_TRANSPORTER_COUNT', 'sum')).reset_index()
```

```python
prosidcounts['sid_cat'] = pd.cut(
    prosidcounts.siderophore_bins, 
    [0,1,3,15], 
    include_lowest=True, 
    right=False, 
    labels=["none", "1-2", "3 or more"])
```

```python
prosidcounts = prosidcounts.rename(columns={'LATITUDE':'lat', 'LONGITUDE':'lon'})
```

### Formatting Synechcocccus counts

```python
synbins = bins[bins['GROUP'].isin(['synechococcus'])]
```

```python
synbincounts = synbins.groupby(['LATITUDE','LONGITUDE']).size().reset_index(name="bins").rename(
    columns={'LATITUDE':'lat', 'LONGITUDE':'lon'})
```

```python
# to make the circle size more visible in map
synbincounts['bins'] = synbincounts['bins'].clip(lower=4)
```

```python
synsidcounts = synbins[['LATITUDE', 'LONGITUDE', 'SIDEROPHORE_TRANSPORTER_COUNT']].groupby(
    ['LATITUDE','LONGITUDE']).agg(
    siderophore_bins=('SIDEROPHORE_TRANSPORTER_COUNT', 'sum')).reset_index()
```

```python
synsidcounts['sid_cat'] = pd.cut(
    synsidcounts.siderophore_bins, 
    [0,1,3,15], 
    include_lowest=True, 
    right=False, 
    labels=["none", "1-2", "3 or more"])
```

```python
synsidcounts = synsidcounts.rename(columns={'LATITUDE':'lat', 'LONGITUDE':'lon'})
```

### Plotting Prochlorococcus count categories on the map

```python
procolors = np.where(prosidcounts["sid_cat"]=='none','#ffffff','-')
procolors[prosidcounts["sid_cat"]=='1-2'] = '#feb24c'
procolors[prosidcounts["sid_cat"]=='3 or more'] = '#f03b20'

procolors
```

```python
fig = plt.figure(figsize=(9,6))
ax = plt.axes(projection=ccrs.Robinson(central_longitude=205))
ax.set_global()
ax.add_feature(cfeature.COASTLINE, edgecolor='#000000', linewidth=0.5)
ax.add_feature(cfeature.LAND, facecolor='#CCCCCC')
#ax.gridlines()
myrast=ds.FeT_darwin_clim.plot(ax=ax,
           transform=ccrs.PlateCarree(),
           vmin=0,
           #vmax=0.0012,
           cmap=cm.viridis,
           cbar_kwargs={'shrink': 0.6})
myscat=ax.scatter(probincounts.lon, probincounts.lat, 
           edgecolor = '#000000',
           c=procolors,
           s=35*np.sqrt(probincounts.bins),
           cmap=cm.rainbow,
           alpha=1,
           marker='o',
           transform=ccrs.PlateCarree())

cbar_ax = fig.add_axes([0.02, 0.35, 0.02, 0.3])
mycbar = fig.colorbar(myrast, cax=cbar_ax)
mycbar.set_label("", rotation=270)

kw = dict(prop="sizes", num=6, func=lambda s: (s/35)**(2))
fig.legend(*myscat.legend_elements(**kw), loc=[0.91, 0.35], title="Num.\ngenomes")

plt.savefig("../figs/pro_sidero_map_full.png", dpi=300, format="png")
```
